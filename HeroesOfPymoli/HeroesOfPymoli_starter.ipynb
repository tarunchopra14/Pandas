```python
#Run basic calculations to obtain number of unique items
unique_item = len(purchase_data['Item Name'].unique())

#calculations to obtain average price
Average_Price = purchase_data["Price"].mean
avg_price = round(Average_Price(),2)

# calculate number of purchases
Number_of_Purchases = len(purchase_data["Purchase ID"].value_counts())

# Total Revenue
total_revenue = purchase_data["Price"].sum()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Here we want results in "$" 
pd.options.display.float_format = '${:.2f}'.format

# Now to calculate purchase count
grouby_Gender = purchase_data.groupby(["Gender"])
grouby_Gender

# Calculating average of groupby gender
grouby_Gender_avg = round(grouby_Gender["Price"].mean(),2)
grouby_Gender_avg

# To calculate purchase count use 'count' function over "Purchase ID".
groupby_gender_purchase_count = grouby_Gender["Purchase ID"].count()
groupby_gender_purchase_count

# To calculate total purchase value
groupby_gender_sum = grouby_Gender["Price"].sum()
groupby_gender_sum

# To calculate average purchase per person
groupby_pur_per_person = groupby_gender_sum /grouby_Gender["SN"].nunique()
groupby_pur_per_person

# Create Data Frame to show output
grouby_Gender_summary = pd.DataFrame(purchase_data.groupby(["Gender"]).nunique())

grouby_Gender_summary["Purchase Count"] = groupby_gender_purchase_count
grouby_Gender_summary["Average Purchase Price"] = grouby_Gender_avg
grouby_Gender_summary["Avg Total Purchase per Person"] = groupby_pur_per_person
grouby_Gender_summary["Total Purchase Value"] = groupby_gender_sum

grouby_Gender_summary



```

```python
# For final results reorganize groupby gender summary.
reorg_grouby_Gender_summary = grouby_Gender_summary[['Purchase Count','Average Purchase Price',
                                                     'Total Purchase Value',
                                                     'Avg Total Purchase per Person']]

reorg_grouby_Gender_summary
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Here we need results in '%' 
pd.options.display.float_format = '{:.2f}%'.format

# Figure out the minimum age
age_min_df= purchase_data["Age"].min()
age_min_df

# Figure out the minimum and maximum views for a TED Talk
age_max_df= purchase_data["Age"].max()
age_max_df

# Create bins in which to place values based on age
age_bins = [0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,45.9]

# Create labels for these bins
age_range = ["<10", "10 to 14", "15 to 19","20 to 24","25 to 29","30 to 34","35 to 39","40+"]


#pd.cut(purchase_data["Age"], age_bins, labels=age_range,include_lowest=True)
# We can append our bins to purchase_data
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], age_bins, labels=age_range,include_lowest=True)
purchase_data

# Group purchase_data by the "Age Ranges" column
grouped_purchase_data = purchase_data.groupby("Age Ranges")
age_demographic = grouped_purchase_data.nunique()
age_demographic

#now rename 'SN' column to 'Total Count'
age_demographic = age_demographic.rename(columns={"SN":"Total Count"})
age_demographic

#Now to calculate 'percentage of players' 
total_count_sum = age_demographic['Total Count'].sum()
total_count_sum

age_demographic['Percentage of Players'] = (age_demographic['Total Count']/total_count_sum)*100
age_demographic

# Now reorganize age demographic
Reorganized_age_demographic = age_demographic[['Total Count','Percentage of Players']]
Reorganized_age_demographic
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Using groupby filter according to 'SN'
top_spender_data = purchase_data.groupby(["SN"])
top_spender_data

# To calculate top spender 
top_spender_price = top_spender_data["Price"].sum()

# Calculate averege price of the top spender,find the mean.
top_spender_avg_price = top_spender_data["Price"].mean()

# Calculate purchase count.
top_spender_purchase_count = top_spender_data["Purchase ID"].count()

# Create summary Data Frame by adding few new columns.
top_spender_data_summary = pd.DataFrame(purchase_data.groupby(["SN"]).nunique())

top_spender_data_summary["Purchase Count"] = top_spender_purchase_count
top_spender_data_summary["Average Purchase Price"] = top_spender_avg_price
top_spender_data_summary["Total Purchase Value"] = top_spender_price

top_spender_data_summary

# Reorganize data and drop unwanted columns
reorg_top_spender_data_summary = top_spender_data_summary[['Purchase Count','Average Purchase Price','Total Purchase Value']]
reorg_top_spender_data_summary                                             


# Finally sort the data in descending order
top_spender_sorted_values = reorg_top_spender_data_summary.sort_values("Total Purchase Value",ascending=False )
top_spender_sorted_values.head()

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Using groupby filter according to 'Item Name' and'Item ID' on purchase data.
most_popular_items = purchase_data.groupby(["Item ID","Item Name"])
most_popular_items

# Calculate most popular item purchase price
most_popular_items_purchase_price = most_popular_items["Price"].sum()

# Find the averege item purchase
most_popular_items_purchase_avg = most_popular_items["Price"].mean()

# Find the most popular purchase items count
most_popular_items_purchase_count = most_popular_items["Purchase ID"].count()

# Create a summary data fram
most_popular_items_summary = pd.DataFrame(purchase_data.groupby(["Item ID","Item Name"]).nunique())

most_popular_items_summary["Purchase Count"] = most_popular_items_purchase_count
most_popular_items_summary["Item Price"] = most_popular_items_purchase_avg
most_popular_items_summary["Total Purchase Value"] = most_popular_items_purchase_price

most_popular_items_summary

# Reorganize DataFrame
reorg_most_popular_items_summary = most_popular_items_summary[['Purchase Count','Item Price','Total Purchase Value']]
                                                     
reorg_most_popular_items_summary

# Sort DataFrame in descending order
most_popular_sorted_values = reorg_most_popular_items_summary.sort_values("Purchase Count",ascending=False )
most_popular_sorted_values.head()
```
